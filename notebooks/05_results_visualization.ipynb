```python
# =========================================================
# CLASS-WISE METRICS COMPARISON
# NO SMOTE vs SMOTE
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import classification_report

# =========================================================
# CLASS LABELS
# =========================================================

class_mapping = {

    0: "Low Bloom Potential\n(< 5 μg/L)",

    1: "Moderate Productivity\n(5–25 μg/L)",

    2: "Elevated Chlorophyll\n(> 25 μg/L)"
}

# =========================================================
# MODEL ORDER
# =========================================================

model_order = [

    "Random Forest",
    "SVM",
    "kNN",
    "XGBoost"
]

# =========================================================
# COLORS
# =========================================================

model_colors = {

    "Random Forest": "#c6dbef",

    "SVM": "#9ecae1",

    "kNN": "#6baed6",

    "XGBoost": "#08519c"
}

# =========================================================
# STYLE
# =========================================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# =========================================================
# FUNCTION
# =========================================================

def plot_classwise_metrics(
    predictions,
    title,
    save_name
):

    # =====================================================
    # EXTRACT METRICS
    # =====================================================

    records = []

    for model_name, y_pred in predictions.items():

        report = classification_report(

            y_test,

            y_pred,

            labels=[0,1,2],

            output_dict=True,

            zero_division=0
        )

        for class_id, class_name in class_mapping.items():

            records.append({

                "Model": model_name,

                "Class": class_name,

                "Precision":
                    report[str(class_id)]["precision"],

                "Recall":
                    report[str(class_id)]["recall"],

                "F1-score":
                    report[str(class_id)]["f1-score"]
            })

    # =====================================================
    # DATAFRAME
    # =====================================================

    classwise_df = pd.DataFrame(records)

    available_models = [

        m for m in model_order

        if m in classwise_df["Model"].unique()
    ]

    # =====================================================
    # METRICS
    # =====================================================

    metrics = [

        "Precision",

        "Recall",

        "F1-score"
    ]

    # =====================================================
    # FIGURE
    # =====================================================

    fig, axes = plt.subplots(

        nrows=1,

        ncols=3,

        figsize=(22,7),

        sharey=True
    )

    # =====================================================
    # PLOT
    # =====================================================

    for ax, metric in zip(axes, metrics):

        pivot_df = classwise_df.pivot(

            index="Class",

            columns="Model",

            values=metric
        )

        pivot_df = pivot_df.reindex(
            columns=available_models
        )

        x = np.arange(len(pivot_df))

        n_models = len(available_models)

        width = 0.8 / n_models

        offsets = np.linspace(

            -0.4 + width/2,

            0.4 - width/2,

            n_models
        )

        # =================================================
        # BARS
        # =================================================

        for offset, model_name in zip(

            offsets,

            available_models
        ):

            bars = ax.bar(

                x + offset,

                pivot_df[model_name],

                width,

                color=model_colors[model_name],

                edgecolor="black",

                linewidth=0.7,

                label=model_name
            )

            # =============================================
            # VALUE LABELS
            # =============================================

            for bar in bars:

                value = bar.get_height()

                ax.text(

                    bar.get_x()
                    + bar.get_width()/2,

                    value + 0.01,

                    f"{value:.3f}",

                    ha="center",

                    va="bottom",

                    fontsize=7
                )

        # =================================================
        # AXIS
        # =================================================

        ax.set_title(

            metric,

            fontsize=14
        )

        ax.set_xticks(x)

        ax.set_xticklabels(

            pivot_df.index,

            fontsize=9
        )

        ax.set_ylim(0,1.05)

        ax.grid(

            axis="y",

            linestyle="--",

            alpha=0.4
        )

        ax.set_axisbelow(True)

        ax.spines["top"].set_visible(False)

        ax.spines["right"].set_visible(False)

    # =====================================================
    # LABELS
    # =====================================================

    axes[0].set_ylabel(

        "Score",

        fontsize=12
    )

    fig.suptitle(

        title,

        fontsize=16,

        y=0.98
    )

    # =====================================================
    # LEGEND
    # =====================================================

    handles, labels = axes[0].get_legend_handles_labels()

    fig.legend(

        handles,

        labels,

        loc="lower center",

        ncol=len(available_models),

        frameon=False,

        fontsize=10,

        bbox_to_anchor=(0.5,-0.02)
    )

    # =====================================================
    # SAVE
    # =====================================================

    plt.tight_layout(

        rect=[0,0.05,1,0.95]
    )

    plt.savefig(

        save_name,

        dpi=300,

        bbox_inches="tight"
    )

    plt.show()

    return classwise_df

# =========================================================
# WITHOUT SMOTE
# =========================================================

print("\nGenerating No-SMOTE figure...")

classwise_no_smote = plot_classwise_metrics(

    predictions_no_smote,

    "Class-wise Performance Comparison (Without SMOTE)",

    "Figure_Classwise_NoSMOTE.png"
)

# =========================================================
# WITH SMOTE
# =========================================================

print("\nGenerating SMOTE figure...")

classwise_smote = plot_classwise_metrics(

    predictions_smote,

    "Class-wise Performance Comparison (With SMOTE)",

    "Figure_Classwise_SMOTE.png"
)

# =========================================================
# SAVE TABLES
# =========================================================

classwise_no_smote.to_csv(

    "Classwise_Metrics_NoSMOTE.csv",

    index=False
)

classwise_smote.to_csv(

    "Classwise_Metrics_SMOTE.csv",

    index=False
)

# =========================================================
# DISPLAY
# =========================================================

print("\nNO SMOTE RESULTS")
display(classwise_no_smote.round(4))

print("\nSMOTE RESULTS")
display(classwise_smote.round(4))

# =========================================================
# FILES
# =========================================================

print("\nSaved Files:")
print("✓ Figure_Classwise_NoSMOTE.png")
print("✓ Figure_Classwise_SMOTE.png")
print("✓ Classwise_Metrics_NoSMOTE.csv")
print("✓ Classwise_Metrics_SMOTE.csv")
```

```python
# =========================================================
# MULTICLASS ROC CURVE COMPARISON
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    roc_curve,
    auc,
    roc_auc_score
)

from sklearn.preprocessing import (
    label_binarize
)

# =========================================================
# STYLE
# =========================================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# =========================================================
# CLASSES
# =========================================================

classes = np.unique(y_test)

y_test_bin = label_binarize(
    y_test,
    classes=classes
)

# =========================================================
# MODEL ORDER
# =========================================================

model_order = [

    "Random Forest",

    "SVM",

    "kNN",

    "XGBoost"
]

# =========================================================
# COLORS
# =========================================================

model_colors = {

    "Random Forest": "#c6dbef",

    "SVM": "#9ecae1",

    "kNN": "#6baed6",

    "XGBoost": "#08519c"
}

# =========================================================
# FUNCTION
# =========================================================

def plot_multiclass_roc(
    fitted_models,
    title,
    save_name
):

    fig, ax = plt.subplots(
        figsize=(9,7)
    )

    auc_table = []

    # =====================================================
    # LOOP MODELS
    # =====================================================

    for model_name in model_order:

        if model_name not in fitted_models:
            continue

        model = fitted_models[model_name]

        # -------------------------------------------------
        # PROBABILITY / SCORE
        # -------------------------------------------------

        if hasattr(model, "predict_proba"):

            y_score = model.predict_proba(
                X_test
            )

        elif hasattr(model, "decision_function"):

            y_score = model.decision_function(
                X_test
            )

        else:

            print(
                f"Skipping {model_name} "
                f"(no probability output)"
            )
            continue

        # -------------------------------------------------
        # MICRO ROC
        # -------------------------------------------------

        fpr, tpr, _ = roc_curve(

            y_test_bin.ravel(),

            y_score.ravel()
        )

        micro_auc = auc(
            fpr,
            tpr
        )

        # -------------------------------------------------
        # MACRO AUC
        # -------------------------------------------------

        macro_auc = roc_auc_score(

            y_test,

            y_score,

            multi_class="ovr",

            average="macro"
        )

        auc_table.append({

            "Model": model_name,

            "Micro AUC": micro_auc,

            "Macro AUC": macro_auc
        })

        # -------------------------------------------------
        # PLOT
        # -------------------------------------------------

        ax.plot(

            fpr,

            tpr,

            linewidth=2.5,

            color=model_colors[model_name],

            label=(
                f"{model_name} "
                f"(Macro={macro_auc:.3f})"
            )
        )

    # =====================================================
    # RANDOM LINE
    # =====================================================

    ax.plot(

        [0,1],

        [0,1],

        linestyle="--",

        color="black",

        linewidth=1.5,

        label="Random Classifier"
    )

    # =====================================================
    # TITLE
    # =====================================================

    ax.set_title(

        title,

        fontsize=16,

        pad=15
    )

    # =====================================================
    # AXES
    # =====================================================

    ax.set_xlabel(
        "False Positive Rate",
        fontsize=12
    )

    ax.set_ylabel(
        "True Positive Rate",
        fontsize=12
    )

    ax.set_xlim(0,1)

    ax.set_ylim(0,1.02)

    # =====================================================
    # GRID
    # =====================================================

    ax.grid(
        linestyle="--",
        alpha=0.4
    )

    # =====================================================
    # LEGEND
    # =====================================================

    ax.legend(
        loc="lower right",
        fontsize=9
    )

    # =====================================================
    # SPINES
    # =====================================================

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    # =====================================================
    # SAVE
    # =====================================================

    plt.tight_layout()

    plt.savefig(
        save_name,
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()

    # =====================================================
    # TABLE
    # =====================================================

    auc_df = pd.DataFrame(
        auc_table
    ).round(4)

    display(
        auc_df.sort_values(
            "Macro AUC",
            ascending=False
        )
    )

    return auc_df

# =========================================================
# NO SMOTE
# =========================================================

auc_no_smote = plot_multiclass_roc(

    fitted_models_no_smote,

    "Multiclass ROC Curve Comparison (Without SMOTE)",

    "Figure_ROC_NoSMOTE.png"
)

# =========================================================
# SMOTE
# =========================================================

auc_smote = plot_multiclass_roc(

    fitted_models_smote,

    "Multiclass ROC Curve Comparison (With SMOTE)",

    "Figure_ROC_SMOTE.png"
)

# =========================================================
# SAVE TABLES
# =========================================================

auc_no_smote.to_csv(
    "ROC_AUC_NoSMOTE.csv",
    index=False
)

auc_smote.to_csv(
    "ROC_AUC_SMOTE.csv",
    index=False
)

print("\nFiles saved:")
print("✓ Figure_ROC_NoSMOTE.png")
print("✓ Figure_ROC_SMOTE.png")
print("✓ ROC_AUC_NoSMOTE.csv")
print("✓ ROC_AUC_SMOTE.csv")
```

```python
# =========================================================
# MULTI-CLASS ROC CURVES + AUC TABLE
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    roc_auc_score,
    roc_curve,
    auc
)

from sklearn.preprocessing import (
    label_binarize
)

# =========================================================
# STYLE
# =========================================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# =========================================================
# CLASS DEFINITIONS
# =========================================================

classes = np.unique(y_test)

class_labels = {

    0: "Low Chlorophyll (<10 μg/L)",

    1: "Moderate Chlorophyll (10–75 μg/L)",

    2: "High Chlorophyll (>75 μg/L)"
}

# =========================================================
# BINARIZE TEST LABELS
# =========================================================

y_test_bin = label_binarize(
    y_test,
    classes=classes
)

# =========================================================
# MODEL ORDER
# =========================================================

model_order = [

    "Random Forest",

    "SVM",

    "kNN",

    "XGBoost"
]

# =========================================================
# ROC COLORS
# =========================================================

roc_colors = {

    0: "#c6dbef",

    1: "#6baed6",

    2: "#08519c"
}

# =========================================================
# FUNCTION
# =========================================================

def multiclass_roc_analysis(

    fitted_models,

    figure_title,

    save_figure,

    save_table
):

    auc_results = []

    available_models = [

        m for m in model_order

        if m in fitted_models
    ]

    # =====================================================
    # FIGURE
    # =====================================================

    fig, axes = plt.subplots(

        nrows=2,

        ncols=2,

        figsize=(14,10)
    )

    axes = axes.flatten()

    # =====================================================
    # LOOP MODELS
    # =====================================================

    for ax, model_name in zip(

        axes,

        available_models
    ):

        model = fitted_models[model_name]

        # -------------------------------------------------
        # PROBABILITY OUTPUT
        # -------------------------------------------------

        if hasattr(model, "predict_proba"):

            y_proba = model.predict_proba(
                X_test
            )

        elif hasattr(model, "decision_function"):

            y_proba = model.decision_function(
                X_test
            )

        else:

            ax.axis("off")
            continue

        # -------------------------------------------------
        # CLASS-WISE AUC
        # -------------------------------------------------

        auc_class_0 = roc_auc_score(

            (y_test == 0).astype(int),

            y_proba[:,0]
        )

        auc_class_1 = roc_auc_score(

            (y_test == 1).astype(int),

            y_proba[:,1]
        )

        auc_class_2 = roc_auc_score(

            (y_test == 2).astype(int),

            y_proba[:,2]
        )

        # -------------------------------------------------
        # MACRO AUC
        # -------------------------------------------------

        macro_auc = roc_auc_score(

            y_test,

            y_proba,

            multi_class="ovr",

            average="macro"
        )

        # -------------------------------------------------
        # SAVE RESULTS
        # -------------------------------------------------

        auc_results.append({

            "Model": model_name,

            "AUC Low Chlorophyll":
                auc_class_0,

            "AUC Moderate Chlorophyll":
                auc_class_1,

            "AUC High Chlorophyll":
                auc_class_2,

            "Macro AUC":
                macro_auc
        })

        # -------------------------------------------------
        # ROC CURVES
        # -------------------------------------------------

        for i in range(len(classes)):

            fpr, tpr, _ = roc_curve(

                y_test_bin[:, i],

                y_proba[:, i]
            )

            roc_auc = auc(
                fpr,
                tpr
            )

            ax.plot(

                fpr,

                tpr,

                color=roc_colors[i],

                linewidth=2,

                label=(
                    f"{class_labels[i]}\n"
                    f"AUC = {roc_auc:.3f}"
                )
            )

        # -------------------------------------------------
        # RANDOM CLASSIFIER
        # -------------------------------------------------

        ax.plot(

            [0,1],

            [0,1],

            "--",

            color="gray",

            linewidth=1.2
        )

        # -------------------------------------------------
        # TITLE
        # -------------------------------------------------

        ax.set_title(

            f"{model_name}\nMacro AUC = {macro_auc:.3f}",

            fontsize=12
        )

        # -------------------------------------------------
        # AXES
        # -------------------------------------------------

        ax.set_xlabel(

            "False Positive Rate",

            fontsize=10
        )

        ax.set_ylabel(

            "True Positive Rate",

            fontsize=10
        )

        ax.set_xlim(0,1)

        ax.set_ylim(0,1.05)

        # -------------------------------------------------
        # GRID
        # -------------------------------------------------

        ax.grid(

            linestyle="--",

            alpha=0.4
        )

        ax.spines["top"].set_visible(False)

        ax.spines["right"].set_visible(False)

        # -------------------------------------------------
        # LEGEND
        # -------------------------------------------------

        ax.legend(

            fontsize=7,

            loc="lower right",

            frameon=True
        )

    # =====================================================
    # HIDE UNUSED AXES
    # =====================================================

    for i in range(

        len(available_models),

        len(axes)
    ):

        axes[i].axis("off")

    # =====================================================
    # OVERALL TITLE
    # =====================================================

    fig.suptitle(

        figure_title,

        fontsize=16,

        y=0.98
    )

    # =====================================================
    # SAVE FIGURE
    # =====================================================

    plt.tight_layout(
        rect=[0,0,1,0.96]
    )

    plt.savefig(

        save_figure,

        dpi=300,

        bbox_inches="tight"
    )

    plt.show()

    # =====================================================
    # AUC TABLE
    # =====================================================

    auc_table = pd.DataFrame(
        auc_results
    )

    auc_table.iloc[:,1:] = (

        auc_table.iloc[:,1:]
        .round(4)
    )

    auc_table = auc_table.sort_values(

        by="Macro AUC",

        ascending=False

    ).reset_index(drop=True)

    # =====================================================
    # SAVE TABLE
    # =====================================================

    auc_table.to_csv(

        save_table,

        index=False
    )

    print("\n=================================================")
    print(figure_title)
    print("=================================================")

    display(auc_table)

    return auc_table

# =========================================================
# WITHOUT SMOTE
# =========================================================

auc_no_smote = multiclass_roc_analysis(

    fitted_models_no_smote,

    "One-vs-Rest ROC Curves (Without SMOTE)",

    "Figure_ROC_NoSMOTE.png",

    "ROC_AUC_Table_NoSMOTE.csv"
)

# =========================================================
# WITH SMOTE
# =========================================================

auc_smote = multiclass_roc_analysis(

    fitted_models_smote,

    "One-vs-Rest ROC Curves (With SMOTE)",

    "Figure_ROC_SMOTE.png",

    "ROC_AUC_Table_SMOTE.csv"
)

print("\nFiles saved:")
print("✓ Figure_ROC_NoSMOTE.png")
print("✓ Figure_ROC_SMOTE.png")
print("✓ ROC_AUC_Table_NoSMOTE.csv")
print("✓ ROC_AUC_Table_SMOTE.csv")
```

```python
# =========================================================
# RANDOM FOREST FEATURE IMPORTANCE
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# STYLE
# =========================================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# =========================================================
# FUNCTION
# =========================================================

def plot_rf_feature_importance(
    fitted_models,
    feature_names,
    title,
    save_name
):

    # =====================================================
    # CHECK MODEL
    # =====================================================

    if "Random Forest" not in fitted_models:
        print("Random Forest model not found.")
        return None

    rf_model = fitted_models["Random Forest"]

    # =====================================================
    # HANDLE PIPELINE / DIRECT MODEL
    # =====================================================

    if hasattr(rf_model, "named_steps"):

        if "model" in rf_model.named_steps:
            rf_estimator = rf_model.named_steps["model"]

        else:
            rf_estimator = list(
                rf_model.named_steps.values()
            )[-1]

    else:

        rf_estimator = rf_model

    # =====================================================
    # CHECK FEATURE IMPORTANCES
    # =====================================================

    if not hasattr(
        rf_estimator,
        "feature_importances_"
    ):
        raise ValueError(
            "Random Forest model does not contain feature_importances_."
        )

    # =====================================================
    # FEATURE IMPORTANCE TABLE
    # =====================================================

    feature_importance = pd.DataFrame({

        "Feature": feature_names,

        "Importance":
            rf_estimator.feature_importances_
    })

    feature_importance = (

        feature_importance

        .sort_values(
            by="Importance",
            ascending=False
        )

        .reset_index(drop=True)
    )

    # =====================================================
    # COLORS
    # =====================================================

    colors = plt.cm.Blues(

        np.linspace(
            0.4,
            0.9,
            len(feature_importance)
        )
    )

    # =====================================================
    # FIGURE
    # =====================================================

    fig, ax = plt.subplots(
        figsize=(10, 7)
    )

    bars = ax.barh(

        feature_importance["Feature"][::-1],

        feature_importance["Importance"][::-1],

        color=colors[::-1],

        edgecolor="black",

        linewidth=0.7
    )

    # =====================================================
    # TITLE
    # =====================================================

    ax.set_title(
        title,
        fontsize=15,
        pad=15
    )

    # =====================================================
    # LABELS
    # =====================================================

    ax.set_xlabel(
        "Importance Score",
        fontsize=12
    )

    ax.set_ylabel(
        "Predictor Variables",
        fontsize=12
    )

    # =====================================================
    # GRID
    # =====================================================

    ax.grid(
        axis="x",
        linestyle="--",
        alpha=0.4
    )

    ax.set_axisbelow(True)

    # =====================================================
    # VALUE LABELS
    # =====================================================

    offset = (
        feature_importance["Importance"].max()
        * 0.02
    )

    for i, value in enumerate(
        feature_importance["Importance"][::-1]
    ):

        ax.text(

            value + offset,

            i,

            f"{value:.4f}",

            va="center",

            fontsize=8
        )

    # =====================================================
    # AXIS LIMIT
    # =====================================================

    ax.set_xlim(

        0,

        feature_importance["Importance"].max()
        * 1.25
    )

    ax.tick_params(
        axis="both",
        labelsize=10
    )

    # =====================================================
    # SPINES
    # =====================================================

    ax.spines["top"].set_visible(False)

    ax.spines["right"].set_visible(False)

    # =====================================================
    # SAVE
    # =====================================================

    plt.tight_layout()

    plt.savefig(
        save_name,
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()

    return feature_importance

# =========================================================
# FEATURE NAMES
# =========================================================

feature_names = [
    "B11",
    "B2",
    "B2_B3",
    "B3",
    "B3_B2",
    "B3_B4",
    "B4",
    "B5",
    "B5_B4",
    "B8",
    "CI",
    "FAI",
    "NDCI",
    "NDRE",
    "NDVI",
    "NDWI"
]

# =========================================================
# WITHOUT SMOTE
# =========================================================

rf_importance_no_smote = plot_rf_feature_importance(

    fitted_models_no_smote,

    feature_names,

    "Random Forest Feature Importance (Without SMOTE)",

    "Figure_RF_Importance_NoSMOTE.png"
)

display(
    rf_importance_no_smote.round(4)
)

rf_importance_no_smote.to_csv(

    "RF_FeatureImportance_NoSMOTE.csv",

    index=False
)

# =========================================================
# WITH SMOTE
# =========================================================

rf_importance_smote = plot_rf_feature_importance(

    fitted_models_smote,

    feature_names,

    "Random Forest Feature Importance (With SMOTE)",

    "Figure_RF_Importance_SMOTE.png"
)

display(
    rf_importance_smote.round(4)
)

rf_importance_smote.to_csv(

    "RF_FeatureImportance_SMOTE.csv",

    index=False
)

# =========================================================
# DONE
# =========================================================

print("\nFiles saved:")
print("✓ Figure_RF_Importance_NoSMOTE.png")
print("✓ Figure_RF_Importance_SMOTE.png")
print("✓ RF_FeatureImportance_NoSMOTE.csv")
print("✓ RF_FeatureImportance_SMOTE.csv")
```

```python
# =========================================================
# LIYUTAN SPATIAL PREDICTION
# XGBOOST (NO SMOTE)
# =========================================================

import os
import joblib
import numpy as np
import pandas as pd
import rasterio
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from matplotlib.colors import BoundaryNorm

# =========================================================
# LOAD MODEL
# =========================================================

model = joblib.load("/content/XGBoost_NoSMOTE.pkl")

print("✓ Model loaded successfully")

# =========================================================
# TRAINING FEATURES
# =========================================================

features = [

    "B11",
    "B2",
    "B2_B3",
    "B3",
    "B3_B2",
    "B3_B4",
    "B4",
    "B5",
    "B5_B4",
    "B8",

    "CI",
    "FAI",
    "NDCI",
    "NDRE",
    "NDVI",
    "NDWI"
]

# =========================================================
# TIFF FILES
# =========================================================

target_files = [

    "/content/Liyutan_Low_HAB_Sentinel2_Composite_2021-12-15.tif",

    "/content/Liyutan_High_HAB_Sentinel2_Composite_2024-08-19.tif"

]

# =========================================================
# STYLE
# =========================================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# =========================================================
# CLASS COLORS
# =========================================================

class_colors = [
    "green",
    "yellow",
    "red"
]

class_labels = [

    "Low Bloom Potential\n(< 5 μg/L)",

    "Moderate Productivity\n(5–25 μg/L)",

    "Elevated Chlorophyll\n(> 25 μg/L)"
]

pred_cmap = ListedColormap(class_colors)

pred_norm = BoundaryNorm(
    [-0.5, 0.5, 1.5, 2.5],
    pred_cmap.N
)

# =========================================================
# HELPER FUNCTION
# =========================================================

def parse_filename(path):

    filename = os.path.basename(path)

    base = os.path.splitext(filename)[0]

    parts = base.split("_")

    reservoir = parts[0]

    period = parts[1]

    date = parts[-1]

    return reservoir, period, date


# =========================================================
# PREDICTION FUNCTION
# =========================================================

def predict_tiff(tiff_path):

    with rasterio.open(tiff_path) as src:

        image = src.read()

        profile = src.profile.copy()

    # -----------------------------------------------------
    # TIFF BAND ORDER
    # -----------------------------------------------------
    #
    # 0  = B2
    # 1  = B3
    # 2  = B4
    # 3  = B5
    # 4  = B8
    # 5  = B11
    # 6  = B2_B3
    # 7  = B3_B2
    # 8  = B3_B4
    # 9  = B5_B4
    # 10 = CI
    # 11 = FAI
    # 12 = NDCI
    # 13 = NDRE
    # 14 = NDVI
    # 15 = NDWI
    #
    # -----------------------------------------------------

    B2    = image[0].astype(np.float32)
    B3    = image[1].astype(np.float32)
    B4    = image[2].astype(np.float32)
    B5    = image[3].astype(np.float32)
    B8    = image[4].astype(np.float32)
    B11   = image[5].astype(np.float32)

    B2_B3 = image[6].astype(np.float32)
    B3_B2 = image[7].astype(np.float32)
    B3_B4 = image[8].astype(np.float32)
    B5_B4 = image[9].astype(np.float32)

    CI    = image[10].astype(np.float32)
    FAI   = image[11].astype(np.float32)
    NDCI  = image[12].astype(np.float32)
    NDRE  = image[13].astype(np.float32)
    NDVI  = image[14].astype(np.float32)
    NDWI  = image[15].astype(np.float32)

    stack = np.stack([

        B11,
        B2,
        B2_B3,
        B3,
        B3_B2,
        B3_B4,
        B4,
        B5,
        B5_B4,
        B8,

        CI,
        FAI,
        NDCI,
        NDRE,
        NDVI,
        NDWI

    ], axis=-1)

    height, width, bands = stack.shape

    X_image = stack.reshape(-1, bands)

    valid_mask = np.isfinite(X_image).all(axis=1)

    X_valid = pd.DataFrame(
        X_image[valid_mask],
        columns=features
    )

    pred_valid = model.predict(X_valid)

    pred_map = np.full(
        height * width,
        np.nan,
        dtype=np.float32
    )

    pred_map[valid_mask] = pred_valid

    pred_map = pred_map.reshape(
        height,
        width
    )

    return pred_map, profile


# =========================================================
# RUN PREDICTIONS
# =========================================================

results = {}

for tif in target_files:

    reservoir, period, date = parse_filename(tif)

    print(f"Processing: {os.path.basename(tif)}")

    pred_map, profile = predict_tiff(tif)

    results[period] = {

        "date": date,

        "pred_map": pred_map,

        "profile": profile
    }

    # =====================================================
    # SAVE GEOTIFF FOR QGIS
    # =====================================================

    output_tif = tif.replace(
        ".tif",
        "_XGBoost_Prediction.tif"
    )

    save_profile = profile.copy()

    save_profile.update(

        dtype=rasterio.uint8,

        count=1,

        nodata=255,

        compress="lzw"
    )

    pred_save = np.where(
        np.isnan(pred_map),
        255,
        pred_map
    ).astype(np.uint8)

    with rasterio.open(
        output_tif,
        "w",
        **save_profile
    ) as dst:

        dst.write(pred_save, 1)

    print(f"✓ Saved GeoTIFF: {output_tif}")

# =========================================================
# PLOT RESULTS
# =========================================================

fig, axes = plt.subplots(
    1,
    2,
    figsize=(12, 6)
)

for idx, period in enumerate(["Low", "High"]):

    ax = axes[idx]

    pred_map = results[period]["pred_map"]

    date = results[period]["date"]

    ax.imshow(
        pred_map,
        cmap=pred_cmap,
        norm=pred_norm
    )

    ax.set_title(
        f"{period} HAB Period\n{date}",
        fontsize=13,
        fontweight="bold"
    )

    ax.set_xticks([])
    ax.set_yticks([])

    for spine in ax.spines.values():

        spine.set_visible(True)

        spine.set_linewidth(1)

        spine.set_color("black")

fig.suptitle(
    "Liyutan Reservoir Chlorophyll-a Classification\n(XGBoost Without SMOTE)",
    fontsize=16,
    fontweight="bold"
)

cbar = fig.colorbar(
    plt.cm.ScalarMappable(
        cmap=pred_cmap,
        norm=pred_norm
    ),
    ax=axes,
    orientation="horizontal",
    fraction=0.05,
    pad=0.10
)

cbar.set_ticks([0, 1, 2])

cbar.set_ticklabels(class_labels)

cbar.set_label(
    "Predicted Chlorophyll-a Class",
    fontsize=12
)

plt.tight_layout()

plt.savefig(
    "Liyutan_XGBoost_NoSMOTE_Prediction_Map.png",
    dpi=600,
    bbox_inches="tight"
)

plt.show()

print(
    "\n✓ Saved Figure: Liyutan_XGBoost_NoSMOTE_Prediction_Map.png"
)

print(
    "\n✓ Prediction GeoTIFFs are ready for QGIS."
)
```

```python
# =========================================================
# LIYUTAN SPATIAL PREDICTION
# XGBOOST (NO SMOTE)
# =========================================================

import os
import joblib
import numpy as np
import pandas as pd
import rasterio
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from matplotlib.colors import BoundaryNorm

# =========================================================
# LOAD MODEL
# =========================================================

model = joblib.load("/content/XGBoost_NoSMOTE.pkl")

print("✓ Model loaded successfully")

# =========================================================
# TRAINING FEATURES
# =========================================================

features = [

    "B11",
    "B2",
    "B2_B3",
    "B3",
    "B3_B2",
    "B3_B4",
    "B4",
    "B5",
    "B5_B4",
    "B8",

    "CI",
    "FAI",
    "NDCI",
    "NDRE",
    "NDVI",
    "NDWI"
]

# =========================================================
# TIFF FILES
# =========================================================

target_files = [

    "/content/Mingde_Low_HAB_Sentinel2_Composite_2019-03-13.tif",

    "/content/Mingde_High_HAB_Sentinel2_Composite_2021-10-05.tif"

]

# =========================================================
# STYLE
# =========================================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# =========================================================
# CLASS COLORS
# =========================================================

class_colors = [
    "green",
    "yellow",
    "red"
]

class_labels = [

    "Low Bloom Potential\n(< 5 μg/L)",

    "Moderate Productivity\n(5–25 μg/L)",

    "Elevated Chlorophyll\n(> 25 μg/L)"
]

pred_cmap = ListedColormap(class_colors)

pred_norm = BoundaryNorm(
    [-0.5, 0.5, 1.5, 2.5],
    pred_cmap.N
)

# =========================================================
# HELPER FUNCTION
# =========================================================

def parse_filename(path):

    filename = os.path.basename(path)

    base = os.path.splitext(filename)[0]

    parts = base.split("_")

    reservoir = parts[0]

    period = parts[1]

    date = parts[-1]

    return reservoir, period, date


# =========================================================
# PREDICTION FUNCTION
# =========================================================

def predict_tiff(tiff_path):

    with rasterio.open(tiff_path) as src:

        image = src.read()

        profile = src.profile.copy()

    # -----------------------------------------------------
    # TIFF BAND ORDER
    # -----------------------------------------------------
    #
    # 0  = B2
    # 1  = B3
    # 2  = B4
    # 3  = B5
    # 4  = B8
    # 5  = B11
    # 6  = B2_B3
    # 7  = B3_B2
    # 8  = B3_B4
    # 9  = B5_B4
    # 10 = CI
    # 11 = FAI
    # 12 = NDCI
    # 13 = NDRE
    # 14 = NDVI
    # 15 = NDWI
    #
    # -----------------------------------------------------

    B2    = image[0].astype(np.float32)
    B3    = image[1].astype(np.float32)
    B4    = image[2].astype(np.float32)
    B5    = image[3].astype(np.float32)
    B8    = image[4].astype(np.float32)
    B11   = image[5].astype(np.float32)

    B2_B3 = image[6].astype(np.float32)
    B3_B2 = image[7].astype(np.float32)
    B3_B4 = image[8].astype(np.float32)
    B5_B4 = image[9].astype(np.float32)

    CI    = image[10].astype(np.float32)
    FAI   = image[11].astype(np.float32)
    NDCI  = image[12].astype(np.float32)
    NDRE  = image[13].astype(np.float32)
    NDVI  = image[14].astype(np.float32)
    NDWI  = image[15].astype(np.float32)

    stack = np.stack([

        B11,
        B2,
        B2_B3,
        B3,
        B3_B2,
        B3_B4,
        B4,
        B5,
        B5_B4,
        B8,

        CI,
        FAI,
        NDCI,
        NDRE,
        NDVI,
        NDWI

    ], axis=-1)

    height, width, bands = stack.shape

    X_image = stack.reshape(-1, bands)

    valid_mask = np.isfinite(X_image).all(axis=1)

    X_valid = pd.DataFrame(
        X_image[valid_mask],
        columns=features
    )

    pred_valid = model.predict(X_valid)

    pred_map = np.full(
        height * width,
        np.nan,
        dtype=np.float32
    )

    pred_map[valid_mask] = pred_valid

    pred_map = pred_map.reshape(
        height,
        width
    )

    return pred_map, profile


# =========================================================
# RUN PREDICTIONS
# =========================================================

results = {}

for tif in target_files:

    reservoir, period, date = parse_filename(tif)

    print(f"Processing: {os.path.basename(tif)}")

    pred_map, profile = predict_tiff(tif)

    results[period] = {

        "date": date,

        "pred_map": pred_map,

        "profile": profile
    }

    # =====================================================
    # SAVE GEOTIFF FOR QGIS
    # =====================================================

    output_tif = tif.replace(
        ".tif",
        "_XGBoost_Prediction.tif"
    )

    save_profile = profile.copy()

    save_profile.update(

        dtype=rasterio.uint8,

        count=1,

        nodata=255,

        compress="lzw"
    )

    pred_save = np.where(
        np.isnan(pred_map),
        255,
        pred_map
    ).astype(np.uint8)

    with rasterio.open(
        output_tif,
        "w",
        **save_profile
    ) as dst:

        dst.write(pred_save, 1)

    print(f"✓ Saved GeoTIFF: {output_tif}")

# =========================================================
# PLOT RESULTS
# =========================================================

fig, axes = plt.subplots(
    1,
    2,
    figsize=(12, 6)
)

for idx, period in enumerate(["Low", "High"]):

    ax = axes[idx]

    pred_map = results[period]["pred_map"]

    date = results[period]["date"]

    ax.imshow(
        pred_map,
        cmap=pred_cmap,
        norm=pred_norm
    )

    ax.set_title(
        f"{period} HAB Period\n{date}",
        fontsize=13,
        fontweight="bold"
    )

    ax.set_xticks([])
    ax.set_yticks([])

    for spine in ax.spines.values():

        spine.set_visible(True)

        spine.set_linewidth(1)

        spine.set_color("black")

fig.suptitle(
    "Mingde Reservoir Chlorophyll-a Classification\n(XGBoost Without SMOTE)",
    fontsize=16,
    fontweight="bold"
)

cbar = fig.colorbar(
    plt.cm.ScalarMappable(
        cmap=pred_cmap,
        norm=pred_norm
    ),
    ax=axes,
    orientation="horizontal",
    fraction=0.05,
    pad=0.10
)

cbar.set_ticks([0, 1, 2])

cbar.set_ticklabels(class_labels)

cbar.set_label(
    "Predicted Chlorophyll-a Class",
    fontsize=12
)

plt.tight_layout()

plt.savefig(
    "Mingde_XGBoost_NoSMOTE_Prediction_Map.png",
    dpi=600,
    bbox_inches="tight"
)

plt.show()

print(
    "\n✓ Saved Figure: Liyutan_XGBoost_NoSMOTE_Prediction_Map.png"
)

print(
    "\n✓ Prediction GeoTIFFs are ready for QGIS."
)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.patches import Patch

# ==========================================
# JOURNAL STYLE
# ==========================================

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Liberation Serif"]

# ==========================================
# FUNCTION
# ==========================================

def plot_spatial_prediction(
    reservoir_name,
    low_file,
    high_file,
    save_name
):

    low_img = mpimg.imread(low_file)
    high_img = mpimg.imread(high_file)

    fig, axes = plt.subplots(
        1,
        2,
        figsize=(10, 5)
    )

    # ======================================
    # LOW BLOOM
    # ======================================

    axes[0].imshow(low_img)

    axes[0].set_title(
        "(a) Low Bloom Period",
        fontsize=12,
    )

    axes[0].set_xticks([])
    axes[0].set_yticks([])

    # ======================================
    # HIGH BLOOM
    # ======================================

    axes[1].imshow(high_img)

    axes[1].set_title(
        "(b) High Bloom Period",
        fontsize=12,
    )

    axes[1].set_xticks([])
    axes[1].set_yticks([])

    # ======================================
    # BORDERS
    # ======================================

    for ax in axes:

        for spine in ax.spines.values():

            spine.set_visible(True)
            spine.set_linewidth(0.8)
            spine.set_color("black")

    fig.suptitle(
        f"{reservoir_name} Reservoir",
        fontsize=15,
        y=0.95
    )

    # ======================================
    # LEGEND
    # ======================================

    legend_elements = [
        Patch(facecolor="green", edgecolor="black",
              label="Class 0: Low Bloom Potential (< 10 μg/L)"),

        Patch(facecolor="yellow", edgecolor="black",
              label="Class 1: Moderate Productivity (10–75 μg/L)"),

        Patch(facecolor="red", edgecolor="black",
              label="Class 2: Elevated Chlorophyll (> 75 μg/L)")
    ]

    fig.legend(
        handles=legend_elements,
        loc="lower center",
        ncol=3,
        fontsize=10,
        frameon=True,
        bbox_to_anchor=(0.5, -0.02)
    )

    plt.tight_layout(rect=[0, 0.08, 1, 1])

    plt.savefig(
        save_name,
        dpi=600,
        bbox_inches="tight",
        facecolor="white"
    )

    plt.show()

    print(f"Saved: {save_name}")

# ==========================================
# LIYUTAN
# ==========================================

plot_spatial_prediction(
    reservoir_name="Liyutan",
    low_file="/content/Liyutan_Low.jpeg",
    high_file="/content/Liyutan_High.jpeg",
    save_name="Figure_4_19_Liyutan_Spatial_Prediction.jpeg"
)

# ==========================================
# MINGDE
# ==========================================

plot_spatial_prediction(
    reservoir_name="Mingde",
    low_file="/content/Mingde_Low.jpeg",
    high_file="/content/Mingde_High.jpeg",
    save_name="Figure_4_20_Mingde_Spatial_Prediction.jpeg"
)
```
